```python
import os
import cv2

DATA_DIR = './data'
os.makedirs(DATA_DIR, exist_ok=True)

number_of_classes = 1
dataset_size = 100

cap = cv2.VideoCapture(0)  # Change to 1 or 2 if needed

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

for class_id in range(number_of_classes):
    class_dir = os.path.join(DATA_DIR, str(class_id))
    os.makedirs(class_dir, exist_ok=True)

    print(f'\n Ready to collect for class {class_id}')
    print(' Press "Q" to start capturing')

    # Wait for user to press 'q'
    while True:
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.flip(frame, 1)  # ðŸ”„ Flip the frame

        cv2.putText(frame, f'Ready for Class {class_id} - Press Q to start',
                    (30, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Countdown
    for i in range(3, 0, -1):
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.flip(frame, 1)

        cv2.putText(frame, f'Starting in {i}...', (100, 100),
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 4)
        cv2.imshow('frame', frame)
        cv2.waitKey(1000)

    # Start capture
    counter = 3
    while counter < dataset_size:
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.flip(frame, 1)

        cv2.putText(frame, f'Class {class_id} | Image {counter}/{dataset_size}',
                    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
        cv2.imshow('frame', frame)

        # Save flipped frame
        image_path = os.path.join(class_dir, f'{counter}.jpg')
        cv2.imwrite(image_path, frame)
        counter += 1

        if cv2.waitKey(1) & 0xFF == ord('x'):
            break

print("\nâœ… All class images collected.")
cap.release()
cv2.destroyAllWindows()
```
